## Authentication and initializing the API

```python
import requests
import time
import pandas as pd
```

```python
url = "https://catfact.ninja/breeds"
```

```python
response = requests.get(url)
response.status_code
```

```python
response.json()
```

```python
facts = []

for i in range(10):
    response = requests.get(url)
    data = response.json()
    facts.append(data)
    time.sleep(1)
```
